overall invalid start in job_desc_list

```python
import re

non_letter_starting_descs = []
for i in range(len(data)):
    original_list = data[i]['job_desc_list']

    # filter dengan regex: string tidak diawali huruf
    filtered = [s for s in original_list if re.match(r'^[^a-zA-Z]', s)]

    # tambahkan ke list hasil
    non_letter_starting_descs.append(filtered)

    # tampilkan hanya jika hasil tidak kosong
    if filtered:
        print(f"Data ke-{i}:")
        for f in filtered:
            print(f)
        print()
```
