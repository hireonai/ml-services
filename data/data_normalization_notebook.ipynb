```python
import pandas as pd
import numpy as np
```

# Read Data

```python
cleaned_data = pd.read_csv('cleaned_data/merged_full_clean_data.csv').drop('jobs_id', axis=1)
cleaned_data['company_desc'] = cleaned_data['company_desc'].fillna('-')
cleaned_data['salary'] = cleaned_data['salary'].replace('Unpaid', 'Rp0 – 0')
cleaned_data.head()
```

```python
cleaned_data.info()
```

# Data Normalization

## Company Table

```python
companies = cleaned_data[['company_name', 'company_desc', 'company_profile_src', 'company_location', 'company_industry', 'company_employess_count']].drop_duplicates()
companies['company_id'] = companies.reset_index().index
companies.info()
companies.head()
```

## Company Industry

```python
company_industries = cleaned_data[['company_industry']].drop_duplicates()
company_industries['company_industry_id'] = company_industries.reset_index().index
company_industries.info()
company_industries.head()
```

## Job Min Experiences


```python
job_min_experiences = cleaned_data[['min_experience']].drop_duplicates()
job_min_experiences['job_min_experience_id'] = job_min_experiences.reset_index().index
job_min_experiences.info()
job_min_experiences.head()

```

## Jobs

```python
jobs = cleaned_data[['kategori', 'url', 'job_position', 'employment_type', 'working_location_type', 'working_location', 'min_experience', 'salary', 'job_desc_list', 'job_qualification_list', 'company_name']]
jobs['jobs_id'] = jobs.reset_index().index
jobs['salary'] = jobs['salary'].str.replace('Rp', '').str.replace('.', '')
jobs['min_salary'] = jobs['salary'].str.split('–', expand=True)[0]
jobs['max_salary'] = jobs['salary'].str.split('–', expand=True)[1]
jobs = jobs.fillna(0)
jobs.info()
jobs.head()
```

## Job Categories

```python
import ast

# Extract individual categories from the multilabel lists
all_categories = []
for cat_list in cleaned_data['kategori']:
    # Convert string representation of list to actual list
    categories = ast.literal_eval(cat_list)
    all_categories.extend(categories)

# Create a DataFrame with unique categories
unique_job_categories = pd.DataFrame({'category_name': list(set(all_categories))})

# Sort by category name for better readability (optional)
unique_job_categories = unique_job_categories.sort_values('category_name').reset_index(drop=True)

unique_job_categories['job_category_id'] = unique_job_categories.reset_index().index

unique_job_categories.info()
# Display the result
unique_job_categories.head()
```

## Cek

```python
companies.info()
print("--------------------------------")
jobs.info()
print("--------------------------------")
unique_job_categories.info()
print("--------------------------------")
job_min_experiences.info()
print("--------------------------------")
company_industries.info()
print("--------------------------------")
```

## Save to json

```python
import json

# List of dataframes to save
dataframes = {
    'companies': companies,
    'jobs': jobs,
    'unique_job_categories': unique_job_categories,
    'job_min_experiences': job_min_experiences,
    'company_industries': company_industries
}

# Save each dataframe as JSON
for name, df in dataframes.items():
    # Convert to list of dictionaries
    records = df.to_dict(orient='records')
    
    # Save with proper encoding
    with open(f'cleaned_data/final_json/{name}.json', 'w', encoding='utf-8') as f:
        json.dump(records, f, ensure_ascii=False, indent=4)
    
    print(f"Saved {name}.json successfully")
```
