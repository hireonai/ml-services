```python
import pandas as pd
import numpy as np
```

```python
cleaned_data = pd.read_csv('cleaned_data/merged_full_clean_data.csv').drop('jobs_id', axis=1)
cleaned_data.head()
```

```python
temp_data = cleaned_data.drop(['url', 'company_profile_src'], axis=1)
temp_data.head()
```

```python
temp_data['salary'] = temp_data['salary'].str.replace('Negotiable', '')
temp_data['jobs_id'] = temp_data.index
```

```python
temp_data.head()
```

```python
# Membagi data menjadi chunk berukuran 500 baris dan menyimpannya
chunk_size = 500
total_chunks = (len(cleaned_data) + chunk_size - 1) // chunk_size  # Menghitung jumlah chunk yang dibutuhkan

for i in range(total_chunks):
    start_idx = i * chunk_size
    end_idx = min((i + 1) * chunk_size, len(cleaned_data))
    
    # Mengambil chunk data
    chunk = temp_data.iloc[start_idx:end_idx]
    
    # Menyimpan chunk ke file CSV
    output_filename = f'temp_data_chunk_{i+1}_of_{total_chunks}.csv'
    chunk.to_csv(output_filename, index=False)
    
    print(f'Saved {output_filename} with {len(chunk)} rows')
```
